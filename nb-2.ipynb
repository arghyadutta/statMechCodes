Consider two boxes. The first contains three red and four green balls, and the second contains five red and nine green balls. We flip a fair coin and then draw a ball from the first box or the second box depending on whether the outcome was heads or tails. What is the conditional probability that the outcome of the toss was heads given that a red ball was selected?

```python
import random
import matplotlib.pyplot as plt

box1 = ['red'] * 3 + ['green'] * 4
box2 = ['red'] * 5 + ['green'] * 9
steps = [100, 500, 1000, 5000, 10000, 50000, 100000, 500000, 1000000]
probs = []

for n in steps:
    rh = 0  # red from heads
    tr = 0  # total red drawn
    
    for _ in range(n):
        c = random.choice(['H', 'T'])

        if c == 'H':
            b = random.choice(box1)
        else:
            b = random.choice(box2)

        if b == 'red':
            tr += 1
            if c == 'H':
                rh += 1

    if tr != 0:
        prob = rh / tr
    else:
        prob = 0

    probs.append(prob)

plt.plot(steps, probs, 'o-', label='Simulated')
plt.axhline(6/11, color='r', ls='--', label='Theoretical 6/11')
plt.xscale('log')
plt.xlabel('Trials')
plt.ylabel('$P(H|R)$')
plt.legend()
plt.grid();
```
